```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook
```

```python
# Read in csv file
county = pd.read_csv('data/county_facts.csv')
county
```

```python
# Read in csv file
primary = pd.read_csv('data/primary_results.csv')
primary
```

```python
# Descriptive statistics for votes
primary['votes'].describe()
```

```python
# Find total number of votes in each state
state_votes = pd.DataFrame(primary.groupby('state')['votes'].sum())
state_votes
```

```python
# state_votes['votes'].len()
```

```python
# Create bar chart of total votes from each state
state_chart = state_votes.plot(kind='bar')
state_chart.set_xlabel('State')
state_chart.set_ylabel('Number of Votes')
plt.title('Total Votes From Each State')
# plt.xlim(0, 60)
plt.show()
```

```python
# Find total number of votes for each candidate
candidate_votes = pd.DataFrame(primary.groupby('candidate')['votes'].sum())
candidate_votes
```

```python
# Create bar chart for total votes for each candidate
candidate_chart = candidate_votes.plot(kind='bar')
candidate_chart.set_xlabel('Candidate')
candidate_chart.set_ylabel('Number of Votes')
plt.title('Total Votes For Each Candidate')
plt.show()
```

```python
party_votes = pd.DataFrame(primary.groupby('party')['votes'].sum())
party_votes
```

```python
state_votes = primary.groupby(['state', 'party'])['votes'].mean().round(2)
state_votes
```
