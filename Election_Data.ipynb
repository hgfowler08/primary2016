```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook
```

```python
primary = pd.read_csv('data/primary_results.csv')
primary
```

```python
primary['votes'].describe()
```

```python
state_votes = pd.DataFrame(primary.groupby('state')['votes'].sum())
state_votes
```

```python
# state_votes['votes'].len()
```

```python
state_chart = state_votes.plot(kind='bar')
state_chart.set_xlabel('State')
state_chart.set_ylabel('Number of Votes')
plt.title('Total Votes From Each State')
# plt.xlim(0, 60)
plt.show()
```

```python
candidate_votes = pd.DataFrame(primary.groupby('candidate')['votes'].sum())
candidate_votes
```

```python
candidate_chart = candidate_votes.plot(kind='bar')
candidate_chart.set_xlabel('Candidate')
candidate_chart.set_ylabel('Number of Votes')
```
